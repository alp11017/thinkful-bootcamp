```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = 'AB_NYC_2019.csv'
df = pd.read_csv(file)
```

```python
print(df.columns)
#print(df.head())
```

    Choose one variable and plot that variable four different ways.


```python
ax = sns.lmplot(x='latitude', y='longitude', data=df)
plt.show()

ax = sns.lmplot(x='latitude', y='price', data=df)
plt.show()

ax = sns.lmplot(x='latitude', y='minimum_nights', data=df)
plt.show()

ax = sns.lmplot(x='latitude', y='reviews_per_month', data=df)
plt.show()
```

    Choose two continuous variables, and plot them three different ways.


```python
ax = sns.lineplot(x='reviews_per_month', y='price', data=df)
plt.show()

ax = sns.relplot(x='reviews_per_month', y='price', data=df)
plt.show()

ax = sns.jointplot(x='reviews_per_month', y='price', data=df)
plt.show()
```

    Choose one continuous variable and one categorical variable, and plot them six different ways.


```python
ax = sns.stripplot(x='neighbourhood_group', y='price', data=df, jitter=True)
plt.show()

ax = sns.boxplot(x='neighbourhood_group', y='price', data=df)
plt.show()

ax = sns.violinplot(x='neighbourhood_group', y='price', data=df)
plt.show()

ax = sns.boxenplot(x='neighbourhood_group', y='price', data=df)
plt.show()

ax = sns.pointplot(x='neighbourhood_group', y='price', data=df)
plt.show()

ax = sns.catplot(x='neighbourhood_group', y='price', data=df)
plt.show()
```

The violinplot gives us information about the probability density functions, although for this dataset we can see that the graph itself is heavily influenced by very extreme values. We can see that the distributions do in fact vary between the groups. This plot is also useful for comparing the extent of extreme values present within each neighbourhood group
